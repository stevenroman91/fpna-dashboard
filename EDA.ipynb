```python
import pandas as pd
```

## EDA

Data upload

```python
df_client_dim = pd.read_csv('./Data/client_dimension.csv')
df_fpa_actual = pd.read_excel('./Data/fpa_actual.xlsx')
df_fpa_budget = pd.read_excel('./Data/fpa_budget.xlsx')
df_fpa_forecast = pd.read_excel('./Data/fpa_forecast.xlsx')
```

```python
df_client_dim['Join Year'] = pd.to_datetime(df_client_dim['Join Year'], format='%Y')
display(df_client_dim)
```

```python
import pandas as pd

# Supposons que ton DataFrame s'appelle df
df_fpa_actual['Date'] = pd.to_datetime(df_fpa_actual[['Year', 'Month']].assign(DAY=1))
df_fpa_actual.drop(columns=['Year', 'Month'],inplace=True)
df_fpa_actual['Scenario'] = "Actual"
df_fpa_actual.rename(columns={'Subcatego': 'Subcategory'}, inplace=True)

display(df_fpa_actual)
```

```python
df_fpa_budget['Date'] = pd.to_datetime(df_fpa_budget[['Year', 'Month']].assign(DAY=1))
df_fpa_budget.drop(columns=['Year', 'Month'],inplace=True)
df_fpa_budget['Scenario'] = "Budget"
display(df_fpa_budget)
```

```python
df_fpa_forecast['Date'] = pd.to_datetime(df_fpa_forecast[['Year', 'Month']].assign(DAY=1))
df_fpa_forecast.drop(columns=['Year', 'Month'],inplace=True)
df_fpa_forecast['Scenario'] = "Forecast"
df_fpa_forecast.rename(columns={'Count': 'Country'}, inplace=True)

display(df_fpa_forecast)
```

```python
df_fact = pd.concat([df_fpa_actual, df_fpa_budget, df_fpa_forecast], ignore_index=True)
display(df_fact)
```

Export des données pour BI

```python
df_client_dim.to_excel('./Data/final_client_dimension.xlsx')
```

```python
df_fact.to_excel('./Data/df_fact.xlsx')
```
